```python
import pandas as pd
from matplotlib import pyplot as plt

from generate_data import generate_time_series
from plot.new_plot import plot_ts
```

```python
# Example usage
start_time = '2024-01-01 00:00:00'
interval = 1
num_steps = 780
interval_type = 'days'  # Can be 'seconds', 'minutes', 'hours', 'days', 'weeks'

df = generate_time_series(start_time, interval, num_steps, interval_type)
x = df['timestamp']
y = df['value']
x = pd.to_datetime(x)
```

```python
# test plot
fig, ax, new_x = plot_ts(
    x_values=x,
    y_values=y,
)
ax.plot(new_x, y)
plt.show()
```
